```python
import os
import pandas as pd
```

```python
def split_clusters(input_file, output_dir=r"D:\Thesis EGFR dataset\Dataset\Clustering\90 percent\cdrs_clusters_90"):
    os.makedirs(output_dir, exist_ok=True)

    df = pd.read_csv(input_file)

    for cluster_id, group in df.groupby('Cluster_number'):
        out_path = os.path.join(output_dir, f"cluster_{cluster_id}.fasta")
        with open(out_path, "w") as f:
            for _, row in group.iterrows():
                nb_id = row['Nanobody_id']
                label = row['Label']
                seq = row['CDR_sequence']
                header = f"{nb_id}|Label={label}"
                
                f.write(f">{header}\n{seq}\n")

split_clusters(r"D:\Thesis EGFR dataset\Dataset\Clustering\90 percent\cdhit_cluster_list90.csv")
```
