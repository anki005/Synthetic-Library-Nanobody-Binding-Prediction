```python
import pandas as pd
```

```python
cdr3_cdr = pd.read_csv(r"D:\Thesis EGFR dataset\Dataset\CDR3 sequence.csv")
all_cdr = pd.read_csv(r"D:\Thesis EGFR dataset\Dataset\new_filtered_seqs_merged.csv")
```

```python
count = (all_cdr["CDR3"].str.len() < 10).sum()
print(count)
```

```python
all_cdr = all_cdr[all_cdr["CDR3"].str.len() >= 10]
print(len(all_cdr))
```

```python
concat_cdrs = all_cdr[all_cdr["Nanobody_id"].isin(cdr3_cdr["Nanobody_id"])]
concat_cdrs = concat_cdrs[["Nanobody_id", "CDR1", "CDR2", "CDR3", "Label"]]
print(concat_cdrs.head())

concat_cdrs["Concatenated_CDRs"] = concat_cdrs["CDR1"] + concat_cdrs["CDR2"] + concat_cdrs["CDR3"]
concat_cdrs = concat_cdrs[["Nanobody_id", "Concatenated_CDRs", "Label"]]
print(concat_cdrs.head())
```

```python
print(len(concat_cdrs))
concat_cdrs.to_csv(r"D:\Thesis EGFR dataset\Dataset\concatenated_cdrs.csv", index=False)
```
