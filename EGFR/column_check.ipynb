Check the sequences with insertions

```python
check_df = pd.read_csv(r"D:\Thesis - November\Dataset\ANARCII\nanobodies_imgt_wide_fixed.csv")

# columns you want to check
positions = ["6B", "6C", "6D", "9B", "9C", "9D", "9E", "42B", "42C", "42D", "51B"]    # just change this list

output_rows = []
for _, row in check_df.iterrows():
    nb_id = row["Name"]
    inserts = []

    for p in positions:
        col = str(p)      # column name is a string
        if col in check_df.columns:
            aa = row[col]
            if aa not in ["", "-", None]:   # means there is an insertion/residue
                inserts.append(col)

    if inserts:
        output_rows.append({"Name": nb_id, "Insertions": ", ".join(inserts)})

output_df = pd.DataFrame(output_rows)
output_df.to_csv(r"D:\Thesis - November\Dataset\ANARCII\insertions.csv", index=False)
```

Remove the sequences with insertions and delete respective column

```python
insertions = pd.read_csv(r"D:\Thesis - November\Dataset\ANARCII\insertions.csv")
check_df = pd.read_csv(r"D:\Thesis - November\Dataset\ANARCII\nanobodies_imgt_wide_fixed.csv")
df_clean = check_df[~check_df["Name"].isin(insertions["Name"])]

df_clean = df_clean.drop(columns=["6B", "6C", "6D", "9B", "9C", "9D", "9E", "42B", "42C", "42D", "51B"])
df_clean.to_csv(r"D:\Thesis - November\Dataset\ANARCII\nanobodies_imgt_filtered.csv", index=False)
```
