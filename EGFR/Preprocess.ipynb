Add Unique ID to the dataset  

```python

def nanobody_id(sequence):
    seq = str(sequence).strip().upper().replace(" ", "")
    hash_hex = hashlib.sha256(seq.encode()).hexdigest()[:12]
    return f"Nb_{hash_hex}"

combined_df = pd.read_csv(r"D:\Thesis - November\Dataset\EG-1unique eg-2-combined.csv")

print(combined_df.columns)

combined_df["Nanobody_id"] = combined_df["igghAaSequence"].apply(nanobody_id)
combined_df.to_csv(r"D:\Thesis - November\Dataset\Combined dataset with ID.csv", index=False)

```
