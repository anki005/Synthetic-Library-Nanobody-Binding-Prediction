```python

wide_csv = r"D:\Thesis EGFR dataset\Dataset\ANARCII\final_nanobodies_imgt_filtered.csv"
df = pd.read_csv(wide_csv)

meta_cols = ["Name", "Chain", "Score", "Query start", "Query end"]
imgt_cols = [c for c in df.columns if c not in meta_cols]

fasta_path = wide_csv.replace(".csv", ".fasta")

with open(fasta_path, "w") as f:
    for _, row in df.iterrows():
        name = row["Name"]
        seq = "".join(str(row[c]) for c in imgt_cols)
        f.write(f">{name}\n{seq}\n")

print("FASTA written to:", fasta_path)
```
