```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv(r"D:\Thesis EGFR dataset\Dataset\Clustering\90 percent\data\train_data.csv")
val_df = pd.read_csv(r"D:\Thesis EGFR dataset\Dataset\Clustering\90 percent\data\val_data.csv")
test_df = pd.read_csv(r"D:\Thesis EGFR dataset\Dataset\Clustering\90 percent\data\test_combined_data.csv")
all_df = pd.read_csv(r"D:\Thesis EGFR dataset\Dataset\new_filtered_aligned_seqs_merged.csv")
```

```python
train_df[["Nanobody_id_clean", "Label_from_id"]] = train_df["Nanobody_id"].str.extract(r'(.*)\|Label=(\d+)')
train_df["Label_from_id"] = train_df["Label_from_id"].astype(int)

merged = train_df.merge(all_df, left_on=["Nanobody_id_clean", "Label_from_id"], right_on=["Nanobody_id", "Label"], how="inner", suffixes=("", "_drop"))
merged = merged.filter(regex='^(?!.*_drop)')
merged = merged.drop(columns=["Nanobody_id", "Label"])
merged = merged.rename(columns={"Nanobody_id_clean": "Nanobody_id", "Label_from_id": "Label", "CDR1 with gaps": "CDR1 aligned", "CDR2 with gaps": "CDR2 aligned", "CDR3 with gaps": "CDR3 aligned"})
#print(merged.head())
print("\n",merged.columns)
merged.to_csv(r"D:\Thesis EGFR dataset\Dataset\Clustering\90 percent\data\train_data.csv", index=False)
```
