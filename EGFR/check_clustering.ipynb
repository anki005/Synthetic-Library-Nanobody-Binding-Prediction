```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Thesis EGFR dataset\Dataset\Clustering\90 percent\cdhit_cluster_list90.csv")
```

```python
seq_cluster_counts = (
    df.groupby("CDR_sequence")["Cluster_number"]
      .nunique()
      .reset_index(name="n_clusters")
)
```

```python
sequences_in_multiple_clusters = seq_cluster_counts[
    seq_cluster_counts["n_clusters"] > 1
]
```

```python
print("Sequences appearing in more than one cluster:")
print(sequences_in_multiple_clusters)
```

```python


# Check if any Nanobody_id appears in more than one cluster
dup_ids = df.groupby("Nanobody_id")["Cluster_number"].nunique()
print("Nanobody IDs in >1 cluster:", (dup_ids > 1).sum())
```
