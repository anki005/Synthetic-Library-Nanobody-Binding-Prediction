```python
import pandas as pd
```

```python

# Read as comma-separated
df = pd.read_csv(r"D:\Thesis - November\Dataset\combined full sequence with id.csv", sep=",")

# Clean column names
df.columns = df.columns.str.strip()

# Write to FASTA
with open(r"D:\Thesis - November\Dataset\nanobodies.fasta", "w") as f:
    for _, row in df.iterrows():
        f.write(f">{row['Nanobody_id']}\n")
        f.write(f"{row['igghAaSequence']}\n")

print("FASTA file created successfully!")
```
